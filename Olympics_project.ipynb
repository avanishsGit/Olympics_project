```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

```

```python
# dataset 
athletes = pd.read_csv('D:/Olympic/athlete_events.csv')
region = pd.read_csv('D:/Olympic/noc_regions.csv')
```

```python
# Top 15 countries participation
top_15_countries = athletes_avanish.Team.value_counts().sort_values(ascending=False).head(15)
top_15_countries
```

```python
# Total number of female athletes in each olympics
female_participants=athletes_avanish[(athletes_avanish.Sex=='F')& (athletes_avanish.Season=='Summer')][['Sex','Year']]
female_participants=female_participants.groupby('Year').count().reset_index()
female_participants.tail()
```

```python
womenOlympics = athletes_avanish[(athletes_avanish.Sex== 'F')&(athletes_avanish.Season == 'Summer')]
```

```python
menOlympics = athletes_avanish[(athletes_avanish.Sex== 'M')&(athletes_avanish.Season == 'Summer')]
```

```python
sns.set(style='darkgrid')
plt.figure(figsize=(20,10))
sns.countplot(x='Year', data=womenOlympics, palette='Set1')
plt.title('Women Participation')
```

```python
part2 = womenOlympics.groupby('Year')['Sex'].value_counts()
part1 = menOlympics.groupby('Year')['Sex'].value_counts()
plt.figure(figsize=(20,10))
part1.loc[:,'M'].plot(color = 'blue')
part2.loc[:,'F'].plot(color = 'orange')
plt.title('Plot for male and female participants over the time ')



```

```python
# Gold medal for USA
goldMedals = athletes_avanish[(athletes_avanish.Team == 'United States')]
silverMedals = athletes_avanish[(athletes_avanish.Team== 'United States')]
bronzeMedals = athletes_avanish[(athletes_avanish.Team == 'United States')]
```

```python

part1 = goldMedals.groupby('Year')['Medal'].value_counts()
part2 = silverMedals.groupby('Year')['Medal'].value_counts()
part3 = bronzeMedals.groupby('Year')['Medal'].value_counts()
plt.figure(figsize=(20,10))
part1.loc[:,'Gold'].plot(color='gold')
part2.loc[:,'Silver'].plot(color='blue')
part3.loc[:,'Bronze'].plot(color='green')

plt.title('        Plot of no of Gold, Silver and Bronze for USA over the time      ')
```

```python
# Gold medal for India
goldMedals = athletes_avanish[(athletes_avanish.Team == 'India')]
silverMedals = athletes_avanish[(athletes_avanish.Team== 'India')]
bronzeMedals = athletes_avanish[(athletes_avanish.Team == 'India')]
```

```python
part1 = goldMedals.groupby('Year')['Medal'].value_counts()
part2 = silverMedals.groupby('Year')['Medal'].value_counts()
part3 = bronzeMedals.groupby('Year')['Medal'].value_counts()
plt.figure(figsize=(20,10))
part1.loc[:,'Gold'].plot(color='gold')
part2.loc[:,'Silver'].plot(color='blue')
part3.loc[:,'Bronze'].plot(color='green')
plt.title('plot of no of Gold, Silver and Bronze for India over the time')
```
